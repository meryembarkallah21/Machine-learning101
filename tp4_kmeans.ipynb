```python
import numpy as np
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
path_to_png_file = "pic.jpg"
import matplotlib.image as mpimg
img = mpimg.imread(path_to_png_file)
```

```python
pixels = [pixel for row in img for pixel in row]

clusterer= KMeans (n_clusters=3, random_state=0).fit(pixels)
print(clusterer.labels_)
print(clusterer.cluster_centers_)


```

```python
def recolor (pixel) :
  cluster = clusterer.predict (np.array([pixel], int))#c
  return clusterer.cluster_centers_[cluster][0] # moye
```

```python
new_img = [[recolor (pixel) for pixel in row] for row in img]
new_img=np.array(new_img,int)
plt.imshow (new_img)
plt.show()
```
